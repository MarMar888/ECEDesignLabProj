```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Final/3D_Print_Jobs_cleaned.csv')
df.head()
```

```python
nums = df[["Material 1 Qty", "Material 2 Qty", "Print Cost", "Print Time (Hours)"]]
scaler = StandardScaler()
scaler.fit(nums)

for i in range(1,5):
  pca = PCA(n_components=i)
  pca.fit(nums);
  data_pca = pca.transform(nums)
  print("There is/are", i, "component(s):", pca.explained_variance_ratio_.sum())
print()

pca = PCA(n_components=1)
pca.fit(nums);
data_pca = pca.transform(nums)
print("The first component is: ", pca.components_[0])

pca = PCA(n_components=2)
pca.fit(nums);
data_pca = pca.transform(nums)
print("The second component is: ", pca.components_[1])
```
